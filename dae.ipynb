```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV data into a Pandas DataFrame
data = pd.read_csv('data/Train.csv')
```

```python
display(data.head())
```

```python
display(data.isnull().sum())
```

```python
display(data.describe())
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='growth_stage')
plt.title('Distribution of Growth Stage')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='season')
plt.title('Data Distribution by Season')
plt.show()
```

```python
mean_extent_damage = data.groupby('growth_stage')['extent'].mean().sort_values()
print(mean_extent_damage)
```

```python
std_extent_damage = data.groupby('growth_stage')['extent'].std().sort_values()
print(std_extent_damage)
```

```python
mean_extent_damage = data.groupby('season')['extent'].mean().sort_values()
print(mean_extent_damage)
```

```python
std_extent_damage = data.groupby('season')['extent'].std().sort_values()
print(std_extent_damage)
```

```python
# Plot the distribution of 'extent' excluding zeros
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='extent', bins=20, kde=True)
plt.title('Distribution of Extent (excluding zeros)')
plt.xlabel('Extent')
plt.ylabel('Frequency')
plt.show()
```
